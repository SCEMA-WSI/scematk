```python
from scematk.colour import rgb_to_grey, grey_to_od
import dask.array as da
import matplotlib.pyplot as plt
```

```python
img = da.random.randint(0, 255, size=(1000, 1000, 3), chunks=(100, 100, 3), dtype='uint8')
plt.imshow(rgb_to_grey(img), cmap='gray')
plt.axis('off')
plt.show()
```
